```python
import pandas as pd
```

```python
df = pd.read_csv('../data/updated_yt_data.csv')
df.head()
```

```python
# dropping un-necessary columns
df = df.drop(['#', 'ER', 'COUNTRY', 'TOPIC OF INFLUENCE'], axis=1)
df = df.dropna()
df.head(10)
```

```python
# spliiting NAME with IDs
df[['NAME', 'ID']] = df['NAME'].str.split('@', expand=True)
df.head(10)
```

```python
df.to_csv('../data/cleaned_indian_yt.csv')
```
